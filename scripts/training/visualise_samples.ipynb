```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import train as t
```

```python

num_points = 32
x_values = np.linspace(0, 31, num_points)
y_values = np.linspace(0, 31, num_points)
z_values = np.linspace(0, 31, num_points)
x_data, y_data, z_data = np.meshgrid(x_values, y_values, z_values)
w_data1, w_data2 = next(t.sample_generator("train", "phosphate")) 

x_data_flat = x_data.flatten()
y_data_flat = y_data.flatten()
z_data_flat = z_data.flatten()
w_data_flat = [w_data1.flatten(), w_data2.flatten()]

for i,j in zip(*w_data_flat):
    print(i,j)

fig, axes = plt.subplots(1, 2, figsize=(16, 8), subplot_kw={'projection': '3d'})

for i, ax in enumerate(axes):
    scatter = ax.scatter(x_data_flat, y_data_flat, z_data_flat, c=w_data_flat[i], s=w_data_flat[i], cmap='viridis', marker='o', alpha=0.5)

    cbar = plt.colorbar(scatter, ax=ax, pad=0.05)
    cbar.set_label('4th Dimension')

    ax.set_title(f'Grid of Data Points in 3D with 4th Dimension (Plot {i+1})')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

plt.tight_layout()
plt.show()
```
