```python
# Import required libraries
import pandas as pd
import os
import json
import matplotlib.pyplot as plt
```

```python
stats = []
key = "wFcorr"
# data_dir = "/vault/NucleoFind-training-pr/unrefined_data_b/"
data_dir = "/vault/NucleoFind-training-pr/unrefined_reserved_data/"

for path in os.scandir(data_dir):
    stats_path = os.path.join(path, "stats.json")
    if not os.path.exists(stats_path): 
        continue
    with open(stats_path, "r") as json_file: 
        data = json.load(json_file)
        pdb = data["pdb"]
        stat = data["cutoffs"][0]["data"][key]
        stats.append((pdb, float(stat)))

print(len(stats))
```

```python
plt.hist(list(zip(*stats))[1], bins=100, color='gray')
plt.show()
```
